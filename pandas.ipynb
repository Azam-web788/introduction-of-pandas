# List comprehension

```python
result["Grade"] = ["A+" if p >= 90
                   else "A" if p >= 80
                   else "B" if p >= 70
                   else "C" if p >= 60
                   else "D" if p >= 50
                   else "F"
                   for p in result.Percentage]
```

```python
result
```

# Lambda functinos

```python
result["Grade1"] = result.Percentage.apply(lambda p:"A+" if p >= 90
                   else "A" if p >= 80
                   else "B" if p >= 70
                   else "C" if p >= 60
                   else "D" if p >= 50
                   else "F")
```

```python
result
```

# Outline function

```python
def Grading_system(p):
        if p >= 90:
            return "A+"
        elif p >= 80:
            return "A"
        elif p >= 70:
            return "B"
        elif p >= 60:
            return "C"
        elif p >= 50:
            return "D"
        else:
            return "F"
result["Grade2"] = result.Percentage.apply(Grading_system)
```
